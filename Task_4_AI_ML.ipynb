```python
# ==================================================
# Task 4: General Health Query Chatbot
# Using Groq API + Prompt Engineering
# ==================================================

# Step 1: Import required library
from groq import Groq

# Step 2: Enter your Groq API Key, Use your own API Key
API_KEY = "api_key"

# Step 3: Initialize Groq Client
client = Groq(api_key=API_KEY)

# Step 4: Define emergency keywords for safety filtering
emergency_keywords = [
    "heart attack",
    "suicide",
    "overdose",
    "severe bleeding",
    "chest pain",
    "difficulty breathing",
    "can't breathe"
]

# Step 5: Display welcome message
print("=" * 40)
print("General Health Query Chatbot")
print("Type 'exit' to stop")
print("=" * 40)

# Step 6: Start chatbot loop
while True:

    # Take input from user
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() == "exit":
        print("Chatbot: Stay healthy. Goodbye!")
        break

    # Step 7: Safety Filter
    if any(keyword in user_input.lower() for keyword in emergency_keywords):
        print("\nChatbot: This may be a medical emergency.")
        print("Please contact a healthcare professional or emergency services immediately.\n")
        continue

    # Step 8: Prompt Engineering
    prompt = f"""
    Act as a helpful medical assistant.

    Guidelines:
    - Provide only general health information.
    - Do not diagnose diseases.
    - Do not prescribe medications.
    - Keep responses friendly and easy to understand.
    - Encourage users to consult a doctor for serious concerns.

    User Question:
    {user_input}
    """

    try:
        # Step 9: Send query to Groq LLM
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            temperature=0.5
        )

        # Step 10: Extract response
        answer = response.choices[0].message.content

        # Step 11: Display chatbot response
        print("\nChatbot:", answer)
        print("-" * 40)

    except Exception as e:
        # Step 12: Error Handling
        print("Error:", e)
```
